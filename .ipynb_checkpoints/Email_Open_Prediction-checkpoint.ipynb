# Fetch all relevant info from exported gmail file

```python
labels=[]
senders=[]
dates=[]
subjects=[]
bodys=[]
email_num=0
```

```python
#     for index, message in zip(range(limit), mbox):
for message in mbox:
    label=''
    sender=''
    date=''
    subject=''
    body=''
    sub=''
    subject=''
    sen=''
    encoding=None

    email_num+=1
    print('Open email #', email_num)

    label= message['X-Gmail-Labels']
    sen= message['from'] 
    date= message['Date']
    sub=message['Subject']
    
# Get the Subject field
    if sub:
        subject, encoding = decode_header(sub)[0]
        if encoding is not None:
            print(encoding)
            try:         
                subject=subject.decode(encoding)
            except:
                subject=''
                print('Non standart encoding', encoding)
                
# Get the From field    
    encoding=None    
    if sen:
        sender, encoding = decode_header(sen)[0]
        if encoding is not None:
            print(encoding)
            try:
                sender= sender.decode(encoding)
            except:
                sender=''
                print('Non standart encoding', encoding)

# Get the email body   
    if message.is_multipart():
        for part in message.walk():
            if part.is_multipart():
                for subpart in part.walk():
                    if subpart.get_content_type() == 'text/plain':
                        body = subpart.get_payload(decode=True)
                    elif part.get_content_type() == 'text/plain':
                        body = part.get_payload(decode=True)
    elif message.get_content_type() == 'text/plain':
        body = message.get_payload(decode=True)


    labels.append(label)
    senders.append(sender)
    dates.append(date)
    subjects.append(subject)
    bodys.append(body)
```

```python
df=pd.DataFrame(zip(labels,senders,dates,subjects,bodys), columns=['Label','From','Date','Subject','Body'])
```

```python
def count_words(list):
    words={}
    for word in list:
        word=word.strip(' .,:;-_')
        if len(word)>0: 
            if word not in words:
                words[word]=1
            else:
                words[word]=words[word]+1
    return words
```

#### Function for calculating the score of an email subject

```python
def Score(subject):
    
    counted=0
    score_sum=0
    score=0
    
    sub_test=str(subject).split()

    for word in sub_test:
        word=word.strip(' .,:;-_')
        if word in opened_unread:
            score_sum= score_sum + opened_unread[word]
            counted+=1
    if counted>0:
        score=score_sum/counted
    return score
```

#### Function for calculating the max score of a word in an email subject

```python
def Score_m(subject):
    
    counted=0
    score=0
    score_max=0
    
    sub_test=str(subject).split()

    for word in sub_test:
        word=word.strip(' .,:;-_')
        if word in opened_unread:
            if abs(opened_unread[word])>abs(score_max):
                score_max=opened_unread[word]
            counted+=1
    
    return score_max
```

### Prediction of opened emails using machine learning

### Option 1 - LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def evaluate_logistic_regression(X_train, y_train, X_test, y_test):
    
    # Creating and training the logistic regression model
    model = LogisticRegression()
    model.fit(X_train, y_train)

    # Making predictions and evaluating the model on the test set
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    class_report = classification_report(y_test, predictions)
    conf_matrix = confusion_matrix(y_test, predictions)

    return {
        "accuracy": accuracy,
        "confusion_matrix": conf_matrix,
        "classification_report": class_report
    }
```

#### Sender only

### Option 2 - K-nearest neighbors

### Find the best k value

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# Assuming X_train, y_train are your training data and labels
k_values = range(1, 21)  # Testing k from 1 to 20
average_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train_Sender_scoreN_ScoreN,y_train, cv=5)  # 5-fold cross-validation
    average_scores.append(np.mean(scores))

# Find the k with the highest average score
optimal_k = k_values[average_scores.index(max(average_scores))]
print("The optimal value of k is:", optimal_k)
```
